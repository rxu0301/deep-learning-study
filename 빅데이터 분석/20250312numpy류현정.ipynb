### 문제

```python
import numpy as np
h = np.array([1.82,2.76,1.69,1.86,1.77,1.73])
w = np.array([86,74,59,95,80,68])
b = h/w**2
print('대상 키 : ',h)
print('대상 체충 : ',w)
print('대상 비만도 : ',b)
```

```python
players = np.array( [ [ 170,76.4 ], [ 183,86.2 ], [ 181,78.5 ], [ 176,80.1 ] ] )
print( "몸무게가 80 이상인 선수 정보 : ", players [ players [ :, 1 ] >= 80 ] )
print( "키가 180 이상인 선수 정보 : ", players [ players[ : , 0 ] >= 180 ] )
```

```python
info = np.array([[151.4, 172.45, 138.65, 177.63, 207.46],
                 [44.64, 163.5, 112.35, 73.55, 97.83]])
h = info[0]
w = info[1]
h_j = ( h < 150) | ( h > 195)
w_j = w >= 14
i_j = np.where( h_j | w_j )[0]
result = i_j.append(list)
print(result)

# 이번 문제 좀 틀린거 같은디..
```
