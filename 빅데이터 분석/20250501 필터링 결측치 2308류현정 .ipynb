```python
menu_data1 = pd.DataFrame(
    [20000, 15000, 12000, 13000, 15000],
    index=['초밥', '초밥', '갈비탕', '갈비탕', '갈비탕'], columns=['가격'])

menu_data2 = pd.DataFrame(
    [12000, 7000, 8000, 9000, 25000],
    index=['갈비탕', '짜장면', '짜장면', '짜장면', '탕수육'], columns=['가격'])
```
