스타벅스 커피 판매량 데이터 정의

```
y 변수에 월별 커피 판매량 데이터를 ndarray 타입으로 할당하세요.
```

````
import numpy as np
import matplotlib.pyplot as plt


y = np.array( [300, 320, 340, 360, 
               380, 400, 420, 440,
               460, 480, 500, 520] )

print(y)
print(type(y))
````

커피 판매량 시각화

```
커피 판매량을 라인 플롯으로 시각화 하세요.

hint: plt.plot()
```

````
import matplotlib.pyplot as plt

plt.plot(y)    # 1월 ~ 12월 (12개)
plt.show()
````

'월'을 표시해 봅니다.

```
1월 부터 12월까지를 x 변수에 ndarray 타입으로 할당하세요.
```

````
y = np.array([300, 320, 340, 360, 
            380, 400, 420, 440, 
            460, 480, 500, 520])


# x 좌표 값으로 사용할 '1월'~'12월' 데이터
x = np.array(['1월', '2월', '3월', '4월', 
            '5월', '6월', '7월', '8월', 
            '9월', '10월', '11월', '12월' ])

````

라인플롯 시각화

```
plt.plot() 함수를 이용하여 월별 커피 판매량 데이터를 라인플롯으로 시각화 하세요.
```

````
plt.plot(x, y)
plt.show()
````

그리드 표시

```
plt.grid(True)
```

````
# grid 표시

plt.grid(True)
plt.plot(x, y)
plt.show()

````

바 플롯 시각화

```
plt.bar()
```

````
plt.bar(x, y)
plt.show()
````

````

plt.grid(True)
plt.bar(x,y)
plt.show()
````
