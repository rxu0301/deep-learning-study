#데이터 인덱싱(데이터 프레임의 값 선택)
- 데이터셋 전체에서 일부 데이터만 선택할때 인덱싱 기법을 활용
- pandas의 loc,iloc 함수를 사용

```python
df = sns.load_dataset('penguins')
df = df[:11].T
df
```

```python
# 문자형 인덱스인 경우에 값을 접근하는 방법
df.loc['bill_length_mm']
```

```python
# 두개 이상의 문자형 인덱스에 접근하는 방법, 결과는 데이터 프레임으로 반환
df.loc[['bill_length_mm','sex']]
```

```python
# 인덱스와 컬럼의 위치를 저장해서 데이터 찾기
df.loc['bill_length_mm',5]
```

```python
# 인덱스의 연속된 구간에 해당하는 복수으 인덱스를 선택하고 싶다
df.loc['species':'bill_length_mm',7]
```

```python
df.loc[['species',"bill_length_mm",'body_mass_g'],5]
```

```python
df = df.T
df
```

```python
# 특정한 컬럼에서 어떠한 수치 이상이거나 이하인 데이터만 추출
df.loc[df['bill_length_mm']>40]
```

```python
df.loc[df['bill_length_mm']>40,['sex']]
```

```python
df.loc[df['flipper_length_mm']==181]
```

```python
# 숫자 외의 object type의 데이터 조건 조회
df.loc[df['sex']=='Male']
```

```python
# 원하는 행과 열을 지정하고 데이터 삽입하기
df.loc[[1,3],['bill_length_mm']] = 50
df.loc[[1,3]]
```

```python
# 인덱스 전체를 기준으로 bill_length_mm 컬럼에 해당하는 값만 10으로 저장
df.loc[ : , 'bill_length_mm'] = 10
df
```

```python
# flipper_length_mm 컬럼에서 190을 초과하는 값을 0으로 변경
df.loc[df['flipper_length_mm'] > 190 ] = 0
df
```
