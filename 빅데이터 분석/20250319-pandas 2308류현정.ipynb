```python
import numpy as np
import pandas as pd

data = np.random.randint(0, 10, 4)  # index 길이와 맞춰서 4개 생성
index = ['a', 'b', 'c', 'd']

s = pd.Series(data=data, index=index, name='series_from_ndarray')
print(s)
```

```python
s['a']
```

```python
data = {'서울':70000,'부산':52000,'대전':35000,'광주':40000,'제주도':28000}
s = pd.Series(data=data,name='series_from_dict')
print(s)
```

```python
city = ['광주','대전','부산','서울','울릉도']
s = pd.Series(data=data,index=city,name='Series_from_dict')
print(s)
```
