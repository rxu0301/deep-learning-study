```python



import matplotlib.pyplot as plt

data_x  = [1, 2, 3, 4]
data_y1 = [2, 1, 3, 5]
data_y2 = [3, 2, 4, 6]
data_y3 = [4, 2, 5, 7]
data_y4 = [5, 4, 6, 8]

# 각 선을 그리고 Line2D 객체를 반환받아 설정 추가
line1, = plt.plot(data_x, data_y1, linestyle='solid', color='g', label='solid', solid_capstyle='butt')
line1.set_solid_joinstyle('miter')  # solid_style과 유사한 효과

line2, = plt.plot(data_x, data_y2, linestyle='dotted', color='r', label='dotted')
line2.set_solid_joinstyle('round')

line3, = plt.plot(data_x, data_y3, linestyle='dashed', color='hotpink', label='dashed')
line3.set_solid_joinstyle('bevel')

line4, = plt.plot(data_x, data_y4, linestyle='dashdot', color='y', label='dashdot')
line4.set_solid_joinstyle('round')

plt.legend(loc='best')
plt.show()
```

```python
# 히스토그램 : 데이터 분포 확인
iris = sns.load_dataset('iris')
iris.head()
```

```python
plt.hist(iris.sepal_length, color='skyblue')
plt.show()
```

```python
plt.hist(iris.sepal_length, color='skyblue',label='sepal_length')
plt.hist(iris.petal_length, color='violet',label='petal_length')
plt.show()
```

```python
fig = plt.figure(figsize=(10,6))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
```

```python
fig = plt.figure(figsize=(10,6))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

ax1.hist((iris.sepal_length,iris.petal_length),histtype='bar')
ax2.hist((iris.sepal_length,iris.petal_length),histtype='barstacked')
ax3.hist((iris.sepal_length,iris.petal_length),histtype='step')
ax4.hist((iris.sepal_length,iris.petal_length),histtype='stepfilled')

ax1.title.set_text('Type: bar')
ax2.title.set_text('Type: barstacked')
ax3.title.set_text('Type: step')
ax4.title.set_text('Type: stepfilled')

plt.tight_layout()

plt.show()
```

```python
# 막대 그래프 : 범주현 데이터의 개수 확인
iris_petal_length_mean = iris.groupby('species').petal_length.mean()
iris_petal_length_mean
```

```python
plt.bar(iris.species.unique(),iris.groupby('species').petal_length.mean(),color=['gold','green','teal'],width=0.95)
plt.show()
```

```python
plt.barh(iris.species.unique(),iris.groupby('species').petal_length.mean(),color=['gold','green','teal'])
plt.show()
```

```python
# 산점도 : 수치형 데이터 상관관계
plt.scatter(x=iris.petal_length,y=iris.petal_width,color='gold')
plt.show()
```

```python
setosa = iris[iris.species=='setosa']
versicolor = iris[iris.species=='versicolor']
virginica = iris[iris.species=='virginica']

plt.scatter(setosa.sepal_length, setosa.petal_width, color='gold', label='setosa')
plt.scatter(versicolor.sepal_length, versicolor.petal_width, color='green', label='versicolor')
plt.scatter(virginica.sepal_length, virginica.petal_width, color='teal', label='skyblue')

plt.show()
```
