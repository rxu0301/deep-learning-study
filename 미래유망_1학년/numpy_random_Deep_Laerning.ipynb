**1. numpy.random.rand**


*   균일 분포에서 난수를 생성함
*   모든 값이 0과 1사이에서 균일하게 분포



```python
# 3*3 행렬 생성 (0~1 사이의 난수)
data_1 = np.random.rand(3,3)
print(f'균일 분포 난수 : \n{ data_1}\n')
print(f'차원 : {np.ndim(data_1)}\n')
print(f'shape : {np.shape(data_1)}\n')
print(f'size : {np.size(data_1)}\n')
print(f'len : {len(data_1)}\n') # 마지막 차원의 갯수
```

```python
# 5*4 행렬 생성 (0~1 사이의 난수)
random_uniform = np.random.rand(5,4)
print(f'균일 분포 난수 : \n{random_uniform}')
```

**2.numpy.random.rand()**
- 정규 분포(평균=0,표준편차=1)에서 난수 생성

```python
# 2*4 행렬 생성 (평균=0 , 표준편차=1) => 평균 = 0이며 데이터 사이의 차이는 1 정도이다.

data_2 = np.random.rand(2,4)
print(f'정규 분포 난수 : \n {data_2}')
```

**3.numpy.random.randint**

**4.numpy.random.choice()**
- 배열에서 임의의 값 샘플링(추첨)

```python
data_4 = [10,20,30,40,50]

# 중복 없이 3개의 값 추첨
random_sample = np.random.choice(data_4,size=3,replace=False)
print(f'랜덤 추첨(중복 없음) : {random_sample}')

# 중복 허용하여 5개 추첨
random_sample_with_replace = np.random.choice(data_4,size=5,replace=True)
print(f'랜덤 추첨(중복 있음) : {random_sample_with_replace}')
```

**5.numpy.random.shuffle()**
- 배열의 순서를 랜덤하게 섞음
- 딥러닝에서는 데이터를 무작위로 섞는데 사용

```python
# 데이터셋 섞기
data_5 = np.array([1,2,3,4,5])
np.random.shuffle(data_5)
print(f'섞인 데이터셋 : {data_5}')
```

**6.numpy.random.permutatin()**
- permutation : 순열 또는 치환 , 순서가 있는 임의의 집합을 다른 순서로 뒤섞는 연산
- 주어진 배열을 랜덤하게 섞은 복사본을 반환
- 원본 데이터를 유지하면서 섞은 데이터를 생성할 때 사용

```python
data_6 = np.array([1,2,3,4,5])

# 배열 랜덤 섞기 ( 원본 데이터 유지 )
permuted_data = np.random.permutation(data_6)
print(f'배열의 랜덤 섞기 데이터 : {permuted_data}')
print(f'원본 데이터 : {data_6}')
```

**7.numpy.random.uniform()**
- 임의의 범위에서 균일 분포를 따르는 난수를 생성
- 균일 분포 ex) 주사위

```python
# 0에서 10 사이의 균일 분포 난수 100개 생성
# 균일 분포 생성 ( 범위 : 0 ~ 10 )

data_7 = np.random.uniform(0,10,size=100)
print(f'균일 분포 (범위 0~10) : \n {data_7}\n')
```

```python
# 히스토그램 시각화

plt.hist(data_7,bins=10,color='skyblue',edgecolor='black',density=True)
plt.title('Uniform Distribution (0 ~10)')
plt.show()
```

**8.numpy.random.normal()**
- 정규 분포(평균=mu,표준편차=sigma)를 따르는 난수를 생성
- 종 모양 : 사람의 키 , 시험 점수

```python
# 평균=5, 표쥰편차=2 인 정규 분포에서 10개의 난수 생성

data_8 = np.random.normal(loc=54,scale=2,size=10)
print(f'정규 분포 난수 (평균=5, 표준편차=2) : \n {data_8}')
```

**균일분포와 정규분포 비교**

```python
# 균일 분포 생성
data_uniform = np.random.uniform(-3,3,size=1000)

# 정규 분포 생성
data_normal = np.random.normal(loc=0,scale=1,size=1000)

# 히스토그램 비교
plt.hist(data_uniform,bins=20,alpha=0.7,label='Uniform',color='skyblue',density=True)
plt.hist(data_normal, bins=20,alpha=0.7,label='Normal',color='lightcoral',density=True)
plt.legend()
plt.title('Uniform vs Normal Distributtion')
plt.show()
```

# reshape

```python
# 원래 배열 (1차원)

array = np.array([10,20,30,40,50])
print(array)
print(('**********'))

# reshape (-1,1) : 1차원 매열을 2차원으로 변환하면서, 열(colum) 개수를 1로 고정
reshaped_array = array.reshape(-1,1)
print(reshaped_array)
```

```python
# 배열 총 크기 : 6개

array = np.array([1,2,3,4,5,6])

# 2행 ( 열 크기 자동 설정 )

print(array.reshape(2,-1))

# 3행 ( 행 크기 자동 설정 )

print(array.reshape(-1,3))
```
