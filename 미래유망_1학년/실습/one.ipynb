```python
import requests
import xml.etree.ElementTree as ET #XML 데이터 파싱(구문 분석)

api_key = "81149823A37BA4DC023676AACA4B63F1"

url = 'https://krdict.korean.go.kr/api/search'

print("단어 검색 프로그램")
print("검색창에 'x' 입력시 프로그램 종료")

while True:
    word = str(input("단어 입력 : "))

    if word == 'X':
        break

    params = {
        'key': api_key,
        'q': word,
        'part': 'word'}

    response = requests.get(url, params=params)

    if response.status_code == 200: #응답 데이터 확인
        root = ET.fromstring(response.content) # XML 데이터 파싱
        definition = root.find('.//definition')

        if definition is not None:
            print(f"\n '단어 : {word}' \n뜻: {definition.text}\n")
        else:
            print(f"\n '{word}'에 대한 뜻을을 찾을 수 없습니다.\n")
    else:
        print(f"\n 서버 응답 코드: {response.status_code}\n")

print("종료")
```
