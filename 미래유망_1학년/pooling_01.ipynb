```python
# my code
import numpy as np

# 4*4 이미지 데이터
image = np.array([[1, 2, 3, 0], [4, 5, 6, 1], [7, 8, 9, 2], [0, 1, 2, 3]])

# 3*3 커널
kernel = np.array([[1, 0, -1], [1, 0, -1], [1, 0, -1]])

# 합성곱 연산
def apply_convolution(image, kernel):
    h, w = image.shape
    kh, kw = kernel.shape
    output = np.zeros((h - kh + 1, w - kw + 1))

    for i in range(h - kh + 1):
        for j in range(w - kw + 1):
            region = image[i : i + kh, j : j + kw]
            output[i, j] = np.sum(region * kernel)

    return output

# Max Pooling
def max_pooling(image, pool_size=2):
    h, w = image.shape
    output = np.zeros((h // pool_size, w // pool_size))

    for i in range(0, h, pool_size):
        for j in range(0, w, pool_size):
            output[i // pool_size, j // pool_size] = np.max(image[i : i + pool_size, j : j + pool_size])

    return output

# 1. 합성곱 적용
convolve_image = apply_convolution(image, kernel)

# 2. Max Pooling 적용
pooling_image = max_pooling(convolve_image)

# 3. 결과 출력
print("Original Image:\n", image)
print("\nConvolved Image:\n", convolve_image)
print("\nPooling Image:\n", pooling_image)
```
