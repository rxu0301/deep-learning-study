# Series (1차원 데이터 구조)

- **Series** : 모든 유형의 데이터를 보유하는 1차원 레이블 배열
- 인덱스(라벨)와 데이터 값으로 구성

```python
# 기본 Series 생성 - 리스트로부터
s1 = pd.Series([1, 3, 5, np.nan, 6, 8])
s1
```

```python
# 인덱스 설정하여 Series 생성
s2 = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])
s2
```

```python
# 딕셔너리로부터 Series 생성
data = {'과자': 100, '음료': 200, '과일': 150}
s3 = pd.Series(data)
s3
```

```python
# Series 인덱싱
print(s2.iloc[0])  # 위치 기반 (명시적)
print(s2['a'])  # 라벨 기반
```

```python
# Series 슬라이싱
print(s2[0:2])  # 인덱스로 슬라이싱
print(s2[['a', 'c']])  # 특정 라벨 선택
```

# DataFrame (2차원 데이터 구조)

- **DataFrame** : 행과 열로 구성된 표나 2차원 배열처럼 데이터를 보관하는 2차원적 데이터 구조
- 여러 Series의 모임

```python
# 딕셔너리로부터 DataFrame 생성
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 28],
    'Score': [85, 90, 78, 92]
})
df
```

```python
# .shape : DataFrame의 크기 (행, 열)
df.shape
```

```python
# .info() : DataFrame의 정보 조회
df.info()
```

```python
# .describe() : 통계 요약
df.describe()
```

```python
# .head() : 처음 n개 행 보기 (기본값 5)
df.head(2)
```

```python
# .tail() : 마지막 n개 행 보기
df.tail(2)
```

# 데이터 선택 및 필터링

```python
# 열 선택하기
print(df['Name'])  # 단일 열
print()
print(df[['Name', 'Age']])  # 다중 열
```

```python
# .loc[] : 라벨 기반 인덱싱
print(df.loc[0])  # 첫 번째 행
print()
print(df.loc[0, 'Name'])  # 특정 셀
```

```python
# .iloc[] : 위치 기반 인덱싱
print(df.iloc[0])  # 첫 번째 행
print()
print(df.iloc[0, 1])  # 첫 번째 행, 두 번째 열
```

```python
# 조건으로 필터링
print(df[df['Age'] > 28])  # 나이가 28보다 큰 행
```

```python
# 여러 조건으로 필터링
print(df[(df['Age'] > 25) & (df['Score'] >= 85)])
```

# 데이터 조작 (Data Manipulation)

```python
# 새로운 열 추가
df['Grade'] = ['A', 'A', 'B', 'A']
df
```

```python
# 열 삭제
df_drop = df.drop('Grade', axis=1)
df_drop
```

```python
# .rename() : 열 이름 변경
df_renamed = df.rename(columns={'Name': 'Student', 'Age': 'Year'})
df_renamed
```

```python
# .sort_values() : 정렬
df_sorted = df.sort_values(by='Score', ascending=False)
df_sorted
```

# 결측값(Missing Data) 처리

```python
# 결측값 포함 DataFrame 생성
df_missing = pd.DataFrame({
    'A': [1, 2, np.nan, 4],
    'B': [5, np.nan, np.nan, 8],
    'C': [9, 10, 11, 12]
})
df_missing
```

```python
# .isnull() / .isna() : 결측값 확인
df_missing.isnull()
```

```python
# .dropna() : 결측값 행 제거
df_dropped = df_missing.dropna()
df_dropped
```

```python
# .fillna() : 결측값 채우기
df_filled = df_missing.fillna(0)
df_filled
```

```python
# 평균값으로 결측값 채우기
df_mean = df_missing.fillna(df_missing.mean())
df_mean
```

# 그룹화 (Grouping) 및 집계 (Aggregation)

```python
# 그룹화 데이터 생성
sales_df = pd.DataFrame({
    'Department': ['Sales', 'Sales', 'IT', 'IT', 'HR', 'HR'],
    'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank'],
    'Salary': [50000, 55000, 60000, 65000, 45000, 48000]
})
sales_df
```

```python
# .groupby() : 그룹화
grouped = sales_df.groupby('Department')
grouped.head()
```

```python
# .sum() : 합계
grouped['Salary'].sum()
```

```python
# .mean() : 평균
grouped['Salary'].mean()
```

```python
# .count() : 개수
grouped['Salary'].count()
```

```python
# .agg() : 여러 집계 함수 동시 적용
grouped['Salary'].agg(['sum', 'mean', 'count', 'min', 'max'])
```

# 데이터 병합 (Merging & Joining)

```python
# 병합할 데이터 생성
df1 = pd.DataFrame({
    'Key': ['A', 'B', 'C'],
    'Value1': [1, 2, 3]
})
df2 = pd.DataFrame({
    'Key': ['A', 'B', 'C'],
    'Value2': [4, 5, 6]
})
df1, df2
```

```python
# .merge() : SQL처럼 병합
merged = pd.merge(df1, df2, on='Key')
merged
```

```python
# .concat() : 연결
df3 = pd.DataFrame({
    'Key': ['D', 'E'],
    'Value1': [4, 5]
})
concatenated = pd.concat([df1, df3], ignore_index=True)
concatenated
```

# 데이터 변환 (Transformation)

```python
# .apply() : 함수 적용
df_nums = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
})
df_nums.apply(lambda x: x * 2)
```

```python
# .map() : 요소별 함수 적용 (pandas 2.1.0+에서 applymap은 deprecated)
df_nums.map(lambda x: x ** 2)
```

```python
# .astype() : 데이터 타입 변환
df_types = pd.DataFrame({
    'Age': ['25', '30', '35'],
    'Salary': ['50000', '60000', '70000']
})
df_types['Age'] = df_types['Age'].astype(int)
df_types['Salary'] = df_types['Salary'].astype(int)
df_types.dtypes
```

# Pivot 및 Reshape

```python
# Pivot 데이터 생성
pivot_df = pd.DataFrame({
    'Month': ['Jan', 'Jan', 'Feb', 'Feb'],
    'Product': ['A', 'B', 'A', 'B'],
    'Sales': [100, 150, 120, 180]
})
pivot_df
```

```python
# .pivot() : 피벗 테이블
pivot = pivot_df.pivot(index='Month', columns='Product', values='Sales')
pivot
```

```python
# .pivot_table() : 집계 포함 피벗
pivot_agg = pd.pivot_table(pivot_df, values='Sales', index='Month', columns='Product', aggfunc='sum')
pivot_agg
```

```python
# .melt() : 와이드 형식을 롱 형식으로 변환
melted = pd.melt(pivot.reset_index(), id_vars=['Month'], var_name='Product', value_name='Sales')
melted
```

# 통계 함수

```python
# 통계 데이터
stats_df = pd.DataFrame({
    'A': [10, 20, 30, 40, 50],
    'B': [5, 15, 25, 35, 45]
})
stats_df
```

```python
# .mean() : 평균
print('평균:')
print(stats_df.mean())
```

```python
# .std() : 표준편차
print('표준편차:')
print(stats_df.std())
```

```python
# .min() / .max() : 최소값 / 최대값
print('최소값:')
print(stats_df.min())
print('최대값:')
print(stats_df.max())
```

```python
# .corr() : 상관계수
print('상관계수:')
print(stats_df.corr())
```

```python
# .cov() : 공분산
print('공분산:')
print(stats_df.cov())
```
