# **CSV 파일 구글 드라이브에 만들기**
- pandas로 데이터 생성후 csv 파일로 저장하기

```python
import numpy as np
import pandas as pd

#데이터 플레임 생성
data = {
    'name': ['Alice','Bob','Charlie','Diana','Eve','Jack','Tom','Emma'],
    'age' : [14,13,15,14,13,16,15,15],
    'grade':['A','B','A','C','B','C','A','A']
        }

df = pd.DataFrame(data)

#CSV파일로 저장
df.to_csv('sample_data_csv',index=False)
```

위 코드에서 index=False는 데이터프레임을 csv파일로 저장할 때 인덱스 열을 포함하지 않도록 저장하는 옵션이다

# **인덱싱과 슬라이싱**
* 열 이름으로 선택 : df['열 이름']

**(2). 라벨(이름) 기반 인덱싱**

```python
# 특정 행 선택 (index 0)
df.loc[0]
```

```python
# 특정 행과 열 선택
df.loc[0,'name']
```

```python
df.loc[0:2 , ['name','age']] # 첫번째 행부터 세번째까지 name과 age 열 선택
```

**(3). 위치(숫자) 기반 인덱싱**

```python
# 특정 위치의 행과 열 선택
df.iloc[0,1]
```

```python
# 여러 행과 열 선택
df.iloc[0:2 , 0:2]
```
