# **그래프 두개 그리기**

```python
dy = np.diff(y)
dy = np.insert(dy,0,0)/0.01
```

```python
plt.figure(figsize=(6,4))
plt.plot(t,y)
plt.plot(t,dy)
plt.grid()
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.title('Example of sinewave')
plt.savefig('results/Matplotlib_13_7.png', dpi=300, bbox_inches='tight')
plt.show()
```

**범례를 적용하기 위해 label를 적용하고 lengend를** **실행했다**

```python
plt.figure(figsize=(6,4))
plt.plot(t,y,label='sin')
plt.plot(t,dy,label='cos')
plt.grid()
plt.legend() # 범례
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.title('Example of sinewave')
plt.savefig('results/Matplotlib_15_8.png', dpi=300, bbox_inches='tight')
plt.show()
```

**그래프 색상** **바꾸기**

```python
plt.figure(figsize=(6,4))
plt.plot(t,y,label='sin')
plt.plot(t,dy,'r',label='cos')
plt.grid()
plt.legend()
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.title('Example of sinewave')
plt.savefig('results/Matplotlib_17_8.png', dpi=300, bbox_inches='tight')
plt.show()
```

**선 굵기 바꾸기**

```python
plt.figure(figsize=(6,4))
plt.plot(t,y,lw = 3,label='sin') # lw = linewidth의 약자
plt.plot(t,dy,'r',label='cos')
plt.grid()
plt.legend()
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.title('Example of sinewave')
plt.savefig('results/Matplotlib_19_8.png', dpi=300, bbox_inches='tight')
plt.show()
```

**각 축별로 그리는 범위 조절하기**

```python
# X축

plt.figure(figsize=(6,4))
plt.plot(t,y,lw = 3,label='sin')
plt.plot(t,dy,'r',label='cos')
plt.grid()
plt.legend()
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.title('Example of sinewave')
plt.xlim(0,3,14)
plt.savefig('results/Matplotlib_21_11.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# y축

plt.figure(figsize=(6,4))
plt.plot(t,y,lw = 3,label='sin')
plt.plot(t,dy,'r',label='cos')
plt.grid()
plt.legend()
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.title('Example of sinewave')
plt.ylim(-1.2,1.2)
plt.savefig('results/Matplotlib_22_11.png', dpi=300, bbox_inches='tight')
plt.show()
```
