# Matplotlib 불러오기 및 한국어 폰트 설정

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib
import platform

# 한국어 폰트 설정
system = platform.system()
if system == 'Windows':
    plt.rcParams['font.family'] = 'Malgun Gothic'
elif system == 'Darwin':  # macOS
    plt.rcParams['font.family'] = 'AppleGothic'
else:  # Linux
    # Linux는 나눔고딕 설치 필요: sudo apt-get install fonts-noto-cjk
    plt.rcParams['font.family'] = 'Noto Sans CJK JP'

# 음수 기호 표시 (마이너스 기호)
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['figure.dpi'] = 100
```

# 기본 플롯 (Line Plot)

```python
# 간단한 선 그래프
x = np.arange(0, 10, 0.5)
y = np.sin(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label='sin(x)', linewidth=2)
plt.xlabel('X축')
plt.ylabel('Y축')
plt.title('사인 함수')
plt.legend()
plt.grid(True)
plt.savefig('results/Matplotlib_3_11.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# 여러 줄 플롯
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(10, 5))
plt.plot(x, y1, label='sin(x)', color='blue', linewidth=2)
plt.plot(x, y2, label='cos(x)', color='red', linewidth=2)
plt.xlabel('X축')
plt.ylabel('Y축')
plt.title('삼각함수 비교')
plt.legend()
plt.grid(True, alpha=0.3)
plt.savefig('results/Matplotlib_4_13.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# 라인 스타일과 마커
x = np.arange(0, 10, 1)
y = x ** 2

plt.figure(figsize=(8, 5))
plt.plot(x, y, marker='o', linestyle='--', color='green', label='y=x²')
plt.xlabel('X축')
plt.ylabel('Y축')
plt.title('라인 스타일 예시')
plt.legend()
plt.savefig('results/Matplotlib_5_10.png', dpi=300, bbox_inches='tight')
plt.show()
```

# 산점도 (Scatter Plot)

```python
# 산점도 기본
x = np.random.randn(100)
y = np.random.randn(100)

plt.figure(figsize=(8, 6))
plt.scatter(x, y, alpha=0.6, s=100, c='blue')
plt.xlabel('X축')
plt.ylabel('Y축')
plt.title('산점도')
plt.savefig('results/Matplotlib_7_9.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# 색상으로 값 표현
x = np.random.randn(100)
y = np.random.randn(100)
colors = np.random.randn(100)
sizes = np.abs(np.random.randn(100)) * 100

plt.figure(figsize=(8, 6))
plt.scatter(x, y, c=colors, s=sizes, alpha=0.6, cmap='viridis')
plt.colorbar(label='색상값')
plt.xlabel('X축')
plt.ylabel('Y축')
plt.title('색상과 크기로 표현한 산점도')
plt.savefig('results/Matplotlib_8_12.png', dpi=300, bbox_inches='tight')
plt.show()
```

# 히스토그램 (Histogram)

```python
# 기본 히스토그램
data = np.random.randn(1000)

plt.figure(figsize=(8, 5))
plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel('값')
plt.ylabel('빈도')
plt.title('정규분포 히스토그램')
plt.grid(axis='y', alpha=0.3)
plt.savefig('results/Matplotlib_10_9.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# 여러 히스토그램 비교
data1 = np.random.randn(1000)
data2 = np.random.randn(1000) + 2

plt.figure(figsize=(8, 5))
plt.hist(data1, bins=30, alpha=0.6, label='데이터1', color='blue')
plt.hist(data2, bins=30, alpha=0.6, label='데이터2', color='red')
plt.xlabel('값')
plt.ylabel('빈도')
plt.title('히스토그램 비교')
plt.legend()
plt.savefig('results/Matplotlib_11_11.png', dpi=300, bbox_inches='tight')
plt.show()
```

# 막대 그래프 (Bar Plot)

```python
# 기본 막대 그래프
categories = ['A', 'B', 'C', 'D', 'E']
values = [10, 24, 36, 18, 7]

plt.figure(figsize=(8, 5))
plt.bar(categories, values, color='coral', edgecolor='black')
plt.xlabel('카테고리')
plt.ylabel('값')
plt.title('막대 그래프')
plt.savefig('results/Matplotlib_13_9.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# 가로 막대 그래프
categories = ['제품A', '제품B', '제품C', '제품D']
values = [100, 150, 120, 200]

plt.figure(figsize=(8, 5))
plt.barh(categories, values, color='lightgreen', edgecolor='black')
plt.xlabel('판매액')
plt.title('가로 막대 그래프')
plt.show()
```

```python
# 그룹 막대 그래프
categories = ['A', 'B', 'C', 'D']
x = np.arange(len(categories))
width = 0.35

values1 = [10, 24, 36, 18]
values2 = [15, 22, 30, 25]

plt.figure(figsize=(8, 5))
plt.bar(x - width/2, values1, width, label='2022', color='blue', alpha=0.8)
plt.bar(x + width/2, values2, width, label='2023', color='orange', alpha=0.8)
plt.xlabel('카테고리')
plt.ylabel('값')
plt.title('그룹 막대 그래프')
plt.xticks(x, categories)
plt.legend()
plt.savefig('results/Matplotlib_15_16.png', dpi=300, bbox_inches='tight')
plt.show()
```

# 박스 플롯 (Box Plot)

```python
# 기본 박스 플롯
data = [np.random.randn(100), np.random.randn(100) + 2, np.random.randn(100) + 1]

plt.figure(figsize=(8, 5))
plt.boxplot(data, labels=['그룹A', '그룹B', '그룹C'])
plt.ylabel('값')
plt.title('박스 플롯')
plt.grid(axis='y', alpha=0.3)
plt.savefig('results/Matplotlib_17_8.png', dpi=300, bbox_inches='tight')
plt.show()
```

# 파이 차트 (Pie Chart)

```python
# 기본 파이 차트
sizes = [30, 25, 20, 25]
labels = ['A', 'B', 'C', 'D']
colors = ['skyblue', 'orange', 'lightgreen', 'lightcoral']

plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('파이 차트')
plt.axis('equal')
plt.savefig('results/Matplotlib_19_9.png', dpi=300, bbox_inches='tight')
plt.show()
```

# 서브플롯 (Subplots)

```python
# 2x2 서브플롯
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# 첫 번째 서브플롯
x = np.arange(0, 10, 0.5)
axes[0, 0].plot(x, np.sin(x))
axes[0, 0].set_title('sin(x)')
axes[0, 0].grid(True)

# 두 번째 서브플롯
axes[0, 1].plot(x, np.cos(x), color='red')
axes[0, 1].set_title('cos(x)')
axes[0, 1].grid(True)

# 세 번째 서브플롯
data = np.random.randn(100)
axes[1, 0].hist(data, bins=20, color='green', alpha=0.7)
axes[1, 0].set_title('히스토그램')

# 네 번째 서브플롯
axes[1, 1].scatter(np.random.randn(100), np.random.randn(100))
axes[1, 1].set_title('산점도')

plt.tight_layout()
plt.show()
```

```python
# 1행 3열 서브플롯
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

x = np.arange(0, 10, 0.1)
axes[0].plot(x, np.sin(x), color='blue')
axes[0].set_title('사인')

axes[1].plot(x, np.cos(x), color='red')
axes[1].set_title('코사인')

axes[2].plot(x, np.tan(x), color='green')
axes[2].set_title('탄젠트')

plt.tight_layout()
plt.show()
```

# 스타일링 (Styling)

```python
# 그래프 스타일 설정
plt.figure(figsize=(8, 5))

# 배경색 설정
ax = plt.gca()
ax.set_facecolor('#f0f0f0')

x = np.arange(0, 10, 0.5)
plt.plot(x, x**2, linewidth=3, color='darkblue', label='y=x²')
plt.plot(x, x, linewidth=2, color='red', linestyle='--', label='y=x')

# 제목, 라벨 폰트 크기 설정
plt.title('스타일 설정 예제', fontsize=16, fontweight='bold')
plt.xlabel('X축', fontsize=12)
plt.ylabel('Y축', fontsize=12)
plt.legend(fontsize=10, loc='upper left')

# X, Y 축 범위 설정
plt.xlim(0, 10)
plt.ylim(0, 100)

plt.grid(True, alpha=0.3)
plt.savefig('results/Matplotlib_26_22.png', dpi=300, bbox_inches='tight')
plt.show()
```

# Figure와 Axes 객체 방식

```python
# Figure와 Axes 생성
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)

x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

ax.plot(x, y, linewidth=2, color='blue')
ax.set_xlabel('X축')
ax.set_ylabel('Y축')
ax.set_title('Figure와 Axes 방식')
ax.grid(True, alpha=0.3)

plt.show()
```

# 2차원 데이터 시각화 (다양한 플롯)

```python
# 영역 그래프
x = np.arange(0, 10, 0.5)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(8, 5))
plt.fill_between(x, y1, alpha=0.3, label='sin(x)', color='blue')
plt.fill_between(x, y2, alpha=0.3, label='cos(x)', color='red')
plt.plot(x, y1, color='blue', linewidth=2)
plt.plot(x, y2, color='red', linewidth=2)
plt.xlabel('X축')
plt.ylabel('Y축')
plt.title('영역 그래프')
plt.legend()
plt.savefig('results/Matplotlib_30_14.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# 등고선 그래프
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

plt.figure(figsize=(8, 6))
contour = plt.contour(X, Y, Z, levels=15, cmap='viridis')
plt.clabel(contour, inline=True, fontsize=8)
plt.colorbar(label='높이')
plt.title('등고선 그래프')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# 3D 기본 설정 (2D 시각화의 예제)
from mpl_toolkits.mplot3d import Axes3D

# 3D 산점도
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

x = np.random.randn(100)
y = np.random.randn(100)
z = np.random.randn(100)

ax.scatter(x, y, z, c='blue', alpha=0.6)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D 산점도')

plt.show()
```

# 범례와 주석

```python
# 주석 추가
plt.figure(figsize=(8, 5))
x = np.arange(0, 10, 0.1)
y = x**2

plt.plot(x, y, linewidth=2, color='blue')

# 특정 점에 주석 추가
plt.annotate('최소점', xy=(0, 0), xytext=(2, 20),
            arrowprops=dict(arrowstyle='->', color='red'),
            fontsize=10)

plt.annotate('최대점', xy=(9.9, 98), xytext=(7, 80),
            arrowprops=dict(arrowstyle='->', color='red'),
            fontsize=10)

plt.xlabel('X축')
plt.ylabel('Y축')
plt.title('주석 예제')
plt.savefig('results/Matplotlib_34_19.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# 범례 위치 설정
plt.figure(figsize=(8, 5))

x = np.linspace(0, 2*np.pi, 100)
plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.cos(x), label='cos(x)')
plt.plot(x, np.tan(x), label='tan(x)')

# 범례 위치 지정
plt.legend(loc='upper left', fontsize=10, framealpha=0.9)
plt.xlabel('X축')
plt.ylabel('Y축')
plt.title('범례 위치 설정')
plt.xlim(0, 2*np.pi)
plt.savefig('results/Matplotlib_35_14.png', dpi=300, bbox_inches='tight')
plt.show()
```

# Pandas와 Matplotlib 통합

```python
# DataFrame으로 플롯
df = pd.DataFrame({
    'Month': ['1월', '2월', '3월', '4월', '5월'],
    'Sales': [100, 150, 120, 200, 180],
    'Profit': [20, 35, 25, 50, 40]
})

plt.figure(figsize=(10, 5))
df.plot(x='Month', y=['Sales', 'Profit'], kind='bar', figsize=(8, 5))
plt.title('월별 판매량과 이익')
plt.ylabel('금액')
plt.tight_layout()
plt.show()
```

```python
# DataFrame 산점도
df_scatter = pd.DataFrame({
    'X': np.random.randn(50),
    'Y': np.random.randn(50)
})

plt.figure(figsize=(8, 6))
plt.scatter(df_scatter['X'], df_scatter['Y'], alpha=0.6, s=100)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('DataFrame 산점도')
plt.savefig('results/Matplotlib_40_11.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# DataFrame 히스토그램
df_hist = pd.DataFrame({
    'Values': np.random.randn(1000)
})

plt.figure(figsize=(8, 5))
df_hist['Values'].hist(bins=30, color='skyblue', edgecolor='black')
plt.title('DataFrame 히스토그램')
plt.xlabel('값')
plt.ylabel('빈도')
plt.show()
```
