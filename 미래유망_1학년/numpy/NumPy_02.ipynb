```python
#리스트를 데이터로 바꾸기

import numpy as np

a = np.array([30,80])
print(a)

print('차원',a.ndim)
print('형태',a.shape)
```

```python
#튜플 데이터 만들기

import numpy as np

a = np.array((30,80))
print(a)

print('차원',a.ndim)
print('형태',a.shape)
```

```python
# 다차원 numpy 만들기

import numpy as np

a = np.array([[30,80],[80,30]])
print(a)

print('차원',a.ndim)
print('형태',a.shape)
```

```python
# 2.2.2의 3차원 numpy 배열

import numpy as np

a = np.array([[[30,80],[80,30]],[[40,20],[60,10]]])
print(a)

print('차원',a.ndim)
print('형태',a.shape)
```
