```python
color_img = cv2.imread('스크린샷 2024-11-04 233508.png') #이미지 읽기 openCV는 BGR순으로 이미지를 읽는다
```

```python
print(f"lamge shape : {color_img.shape}") # 이미지 크기 확인 [출력 : (높이 , 너비 , 채널)]
```

```python
# 특정 픽셀 값 확인
x,y = 100,200 # 픽셀 좌표
b,g,r = color_img[x,y] # 특정 픽셀의 색상 값 (B,G,R채널)
print(f"Pixel value at ({x},{y}) : B={b} , G={g}, R={r}")
```

```python
# 채널 수 확인
channels = color_img.shape[2] # shape[2] 채널 수
print(f'Number of channels : {channels}') # 컬러 이미지라면 3출력
```

```python
rgb_img = cv2.cvtColor(color_img, cv2.COLOR_BGR2RGB)  # BGR -> RGB 변환
plt.imshow(rgb_img)
plt.title('Color Image')
plt.axis('off')
plt.show()
```
