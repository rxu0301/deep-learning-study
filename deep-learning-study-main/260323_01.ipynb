```python
df = pd.read_csv('dataset/winequality-white.csv',sep=';')
df.head()
```

```python
x = df.drop('quality',axis=1)
y = df['quality']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.3,shuffle=True)
```

```python
model =Sequential()
model.add(Dense(256,input_shape=(11,),activation='relu',name='input'))
model.add(Dense(64,activation='relu',name='hidden'))
model.add(Dense(11,activation='softmax',name='output'))
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history = model.fit(X_train,y_train,epochs=100,batch_size=32)
```

# 와인의 분류

```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('dataset/wine.csv',header=None)
df.head()
```

```python
X = df.iloc[:,0:12]
y = df.iloc[:,12]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
```

```python
model =Sequential()
model.add(Dense(30,input_shape=(12,),activation='relu',name='input'))
model.add(Dense(12,activation='relu',name='hidden_01'))
model.add(Dense(8,activation='relu',name='hidden_02'))
model.add(Dense(1,activation='sigmoid',name='output'))
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history = model.fit(X_train,y_train,epochs=100,batch_size=32)
```

```python
loss, acc = model.evaluate(X_test,y_test)
print("정확도 :",acc*100)
```

```python
history = model.fit(X_train, y_train, epochs=2000, batch_size=500,verbose=0,validation_split=0.25)
```

```python
df = pd.DataFrame(history.history)
df
```

```python
y_vloss = df['val_loss']
y_loss = df["loss"]

n = np.arange(len(y_loss))

plt.plot(n, y_vloss, c='red', markersize=2, label='val_loss')
plt.plot(n, y_loss, c='blue', markersize=2, label='train_loss')

plt.legend()
plt.xlabel('epoch')
plt.ylabel('loss')
plt.show()
```

# 학습 중단 Earlystopping

```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
```

```python
df = pd.read_csv('dataset/wine.csv',header=None)

X = df.iloc[:,0:12]
y = df.iloc[:,12]

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
```

```python
model =Sequential()
model.add(Dense(30,input_shape=(12,),activation='relu',name='input'))
model.add(Dense(12,activation='relu',name='hidden_01'))
model.add(Dense(8,activation='relu',name='hidden_02'))
model.add(Dense(1,activation='sigmoid',name='output'))
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model_path = "./test/model_learn_point/16--0.9930.keras"
checkpointer = ModelCheckpoint(filepath=model_path, verbose=0,save_best_only=True,moniter='val_loss')
early_stop = EarlyStopping(monitor='val_loss',patience=20)
history = model.fit(X_train, y_train, epochs=2000, batch_size=500,verbose=1,validation_split=0.25,callbacks=[checkpointer,early_stop])
```
